```python
# Clone pyenv into home directory
git clone https://github.com/pyenv/pyenv.git ~/.pyenv

# Set up environment variables
export PYENV_ROOT="$HOME/.pyenv"
export PATH="$PYENV_ROOT/bin:$PATH"

# Initialize pyenv automatically in shell
eval "$(pyenv init --path)"
eval "$(pyenv init -)"
```

```python
pyenv install 3.10.14   # replace with the desired 3.10.x version
pyenv global 3.10.14    # sets it as default in this shell
```
